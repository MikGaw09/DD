# Zadanie 1

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

#wczytuje dane
zbior = load_breast_cancer()

dane = pd.DataFrame(zbior.data, columns=zbior.feature_names)
target = pd.Series(zbior.target, name="TARGET")

#podstawowe infor
print("wymiar danych:", dane.shape)
print("liczba unikatowych wartosci w wektorze target:", target.nunique())
print()

#obliczam wariancje kolumn
wariancje = dane.var().sort_values()

#wybieram 5 kolumn z najmniejsza wariancja
kolumny_do_usuniecia = wariancje.head(5).index.tolist()

print("Proponowane kolumny do usuniecia:")
for kolumna in kolumny_do_usuniecia:
    print("-", kolumna)

#usuwam kolumny
dane_obciete = dane.drop(columns=kolumny_do_usuniecia)

#dodaje target
nowy_zbior = dane_obciete.copy()
nowy_zbior["TARGET"] = target

print()
print("Nowy wymiar danych:", nowy_zbior.shape)

#zapis do csv
nowy_zbior.to_csv("dataset_cut.csv", sep=";", index=False)

print("Zapisano do pliku dataset_cut.csv")

#podglad danych
display(nowy_zbior.head())
```

Zbiór po tym jak usunąłem 5 kolumn z najmneijsza wariancją ma 26 cech zamiast 30 dzieki czemu dane sa uproszczone.

# Zadanie 4

```python
from sklearn.feature_selection import SelectKBest, f_classif

#wybieram 5 najlepszych cech
wybor_cech = SelectKBest(score_func=f_classif, k=5)
nowe_dane = wybor_cech.fit_transform(dane, target)

#sprawdzam nazwy wybranych kolumn
wybrane_kolumny = dane.columns[wybor_cech.get_support()].tolist()

print("Wybrane kolumny:")
for kolumna in wybrane_kolumny:
    print("-", kolumna)

#tworze nowy zbior danych
wynik = pd.DataFrame(nowe_dane, columns=wybrane_kolumny)
wynik["TARGET"] = target

print()
print("Wymiar danych po redukcji:", wynik.shape)

#zapis do csv
wynik.to_csv("dataset_algorithm.csv", sep=";", index=False)
print("Zapisano plik dataset_algorithm.csv")
print()

#podglad
display(wynik.head())
```

W tym zadaniu wykorzystałem metode SelectKBest która wybiera najlepsze cechy na podstawie ich związku ze zmeinna target. Ustawiłem wybór 5 cech dzieki czemu liczba kolumn w zbiorze została zmniejszona i w zbiorze zostały tylko najbardziej istotne.
